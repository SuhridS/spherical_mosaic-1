```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
def read_data(file, keyword="numOfKeyPoints"):
    data = []
    with open(file, 'r') as f:
        for line in f:
            if keyword in line:
                data.append(line.strip().split(" ")[0])
    return data
```

```python
file = '../log.txt'
created = read_data(file, keyword="created")
updated = read_data(file, keyword="updated")
```

```python
created = [int(k) for k in created]
updated = [int(k) for k in updated]
total = [i + j for i, j in zip(created, updated)]
```

```python

plt.hist(total, bins=20)
plt.ylabel("number of frames")
plt.xlabel("number of inlier matches")
plt.title("number of inlier distribution")
plt.show()
```

```python
plt.hist(created, bins=20, alpha=0.5, label='scenepoints created')
plt.hist(updated, bins=20, alpha=0.5, label='scenepoints updated')
plt.ylabel("number of frames")
plt.xlabel("number of scene points")
plt.title("number of scene points created and updated distribution")
plt.legend()
plt.show()
```
